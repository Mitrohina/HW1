```
import cv2
import numpy as np
import os

X = None
Y = np.array([])
def read_files(X, Y, path, ans):
  files = os.listdir(path)
  for name in files:
    img = cv2.imread(path + '/' + name, 0)
    if img.shape != 0:
      img = cv2.resize(img, (256, 256))
      vect = img.reshape(1, 256 ** 2)
      vect = vect / 255.
      X = vect if (X is None) else np.vstack((X, vect)) 
      Y = np.append(Y, ans)
  return X, Y

def init_model(input_size=256):
    
    ###ЗАДАЧА: проинициализируйте веса модели так, чтобы массив w имел размер (input_size^2,1), а b был числом
    w = 0.0001*np.ones((input_size**2, 1))
    b = 0.0001
    return w, b

def sigmoid(z):
    ###ВАЖНО: функция принимает на вход массив любых размеров, на выход возвращает массив такого же размера
    s = 1/(1 + np.exp(-z))
    return s

def propagate(w, b, X, Y):
    
    m = X.shape[0]
    
#    B = np.empty((m, 1), dtype = float)
#    B[:] = b
    A = sigmoid(np.dot(w.T, X.T) + b)
    print (A)
    summa = 0
    for i in range(m):
        cost = -Y[i]*np.log(A[0][i]) - (1 - Y[i])*np.log(1 - A[0][i])
        summa = summa + cost
    
    errf = summa / m
    dw = 1./m * np.dot(X.T, (A - Y).T)
    summa = 0
    for i in range(m):
        q = A[0][i] - Y[i]
        summa = summa + q
    db = summa / m
    
    
    grads = {"dw": dw,
             "db": db}
    
    return grads, cost

def optimize(w, b, X, Y, num_iterations, learning_rate, print_cost = False):

    
    costs = []
    
    for i in range(num_iterations):
        
        
        ###Напишите значения для градиентов и функции ошибки
        grads, cost = propagate(w, b, X, Y)
        
        
        # Retrieve derivatives from grads
        dw = grads["dw"]
        db = grads["db"]
        
        # обновление параметров
        
        ### START CODE HERE ###
        w = w - learning_rate*dw
        b = b - learning_rate*db
        ### END CODE HERE ###
        
        # Record the costs
        if i % 100 == 0:
            costs.append(cost)
        
        # Print the cost every 100 training iterations
        if print_cost and i % 100 == 0:
            print ("Cost after iteration %i: %f" %(i, cost))
    
    params = {"w": w,
              "b": b}
    
    grads = {"dw": dw,
             "db": db}
    
    return params, grads, costs

def predict(w, b, X):
    
    m = X.shape[0]
    Y_prediction = np.zeros((1,m))
    w = w.reshape(X.shape[1], 1)
    
    
    ### START CODE HERE ### (≈ 1 line of code)
    A = sigmoid(np.dot(w.T, X.T) + b)
    print(A)
    ### END CODE HERE ###
    
    for i in range(A.shape[1]):
        
        # Установите порог, выше которого считаем, что модель выдает 1, а ниже - ноль
        ### START CODE HERE ###
        if A[0][i] > 0.8 : 
            Y_prediction[0][i] = 1
        ### END CODE HERE ###
    
    
    return Y_prediction
  
print('hi')
```

```
w, b = init_model(input_size=256)

grads, cost = propagate(w, b, X_train, Y_train)
num_iterations = 500
learning_rate = 0.01
params, grads, costs = optimize(w, b, X_train, Y_train, num_iterations, learning_rate, print_cost = False)
```

```
w = params['w']
b = params['b']
Y_prediction = predict(w, b, X_test)
for i in range(Y_prediction.shape[1]):
  print (Y_prediction[0][i], Y_test[i])
  if Y_test[i] == Y_prediction[0][i]:
    print ('Correct')
  else:
    print('WRONG!!!')
```
